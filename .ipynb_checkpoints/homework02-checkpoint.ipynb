```python
df = df[columns_to_keep]
```

```python
n_val = int(0.2 * n)
n_test = int(0.2 * n)
n_train = n - (n_val + n_test)
n_val, n_test, n_train, n_val+n_test+n_train
```

```python
idx = np.arange(n)
idx
```

```python
np.random.shuffle(idx)

df_shuffled = df.iloc[idx]
df_shuffled
```

```python
df_train = df_shuffled.iloc[:n_train].copy()
df_val = df_shuffled.iloc[n_train:n_train+n_val].copy()
df_test = df_shuffled.iloc[n_train+n_val:].copy()

df_train.head()
```

```python
df_train = df_train.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)

df_train.head()
```

```python
y_train_orig = df_train.median_house_value.values
y_val_orig = df_val.median_house_value.values
y_test_orig = df_test.median_house_value.values

y_train = np.log1p(df_train.median_house_value.values)
y_val = np.log1p(df_val.median_house_value.values)
y_test = np.log1p(df_test.median_house_value.values)

del df_train['median_house_value']
del df_val['median_house_value']
del df_test['median_house_value']
```

```python
df_train.head()
```

```python
base = ['latitude','longitude','housing_median_age','total_rooms','total_bedrooms',
        'population','households','median_income']

#### METHOD 1: fill it with 0 
def prepare_X1(df):
    df_num = df[base]
    df_num = df_num.fillna(0)
    X = df_num.values
    return X

#### METHOD 2: the mean of this variable
def prepare_X2(df):
    df_num = df[base]
    df_num = df_num.fillna(df['total_bedrooms'].mean())
    X = df_num.values
    return X
```

```python
def train_linear_regression(X, y):
    ones = np.ones(X.shape[0])
    X = np.column_stack([ones, X])

    XTX = X.T.dot(X)
    XTX_inv = np.linalg.inv(XTX)
    w = XTX_inv.dot(X.T).dot(y)
    
    return w[0], w[1:]
```

```python
def rmse(y, y_pred):
    error = y_pred - y
    mse = (error ** 2).mean()
    return np.sqrt(mse)
```

```python
X_train = prepare_X1(df_train)
w_0, w = train_linear_regression(X_train, y_train)

w_0, w
```

```python
X_val = prepare_X1(df_val)
y_pred = w_0 + X_val.dot(w)

y_pred
```

```python
round(rmse(y_val, y_pred),2)
```

```python
X_train = prepare_X2(df_train)
w_0, w = train_linear_regression(X_train, y_train)
X_val = prepare_X2(df_val)
y_pred = w_0 + X_val.dot(w)
round(rmse(y_val, y_pred),2)
```

```python
def train_linear_regression_reg(X, y, r=0.0):
    ones = np.ones(X.shape[0])
    X = np.column_stack([ones, X])

    XTX = X.T.dot(X)
    reg = r * np.eye(XTX.shape[0])
    XTX = XTX + reg

    XTX_inv = np.linalg.inv(XTX)
    w = XTX_inv.dot(X.T).dot(y)
    
    return w[0], w[1:]
```

```python
### For this question, fill the NAs with 0.
for r in [0, 0.000001, 0.0001, 0.001, 0.01, 0.1, 1, 5, 10]:
    X_train = prepare_X1(df_train)
    w_0, w = train_linear_regression_reg(X_train, y_train, r)
    
    X_val = prepare_X1(df_val)
    y_pred = w_0 + X_val.dot(w)
    score = rmse(y_val, y_pred)
    print(r, w_0, round(score,2))
```

```python
result = []
for i in [0,1,2,3,4,5,6,7,9]:
    np.random.seed(i)
    n = len(df)

    n_val = int(0.2 * n)
    n_test = int(0.2 * n)
    n_train = n - (n_val + n_test)

    idx = np.arange(n)
    np.random.shuffle(idx)

    df_shuffled = df.iloc[idx]

    df_train = df_shuffled.iloc[:n_train].copy()
    df_val = df_shuffled.iloc[n_train:n_train+n_val].copy()
    df_test = df_shuffled.iloc[n_train+n_val:].copy()

    df_train = df_train.reset_index(drop=True)
    df_val = df_val.reset_index(drop=True)
    df_test = df_test.reset_index(drop=True)

    y_train_orig = df_train.median_house_value.values
    y_val_orig = df_val.median_house_value.values
    y_test_orig = df_test.median_house_value.values

    y_train = np.log1p(df_train.median_house_value.values)
    y_val = np.log1p(df_val.median_house_value.values)
    y_test = np.log1p(df_test.median_house_value.values)

    del df_train['median_house_value']
    del df_val['median_house_value']
    del df_test['median_house_value']

    X_train = prepare_X1(df_train)
    w_0, w = train_linear_regression(X_train, y_train)
    X_val = prepare_X1(df_val)
    y_pred = w_0 + X_val.dot(w)
    result.append(rmse(y_val, y_pred))
```

```python
result_std = round(np.std(result),3)
result_std
```
